```python
# ============================================================
# EU AI Act Requirements for Providers of High Risk AI Systems
# Interactive Graph Visualisation — Google Colab Version
#
# Based on: AI_Act_Requirements_Graph.ipynb
# Topic:    AI-Based Phishing & Email Security Systems
#           (Proofpoint TAP/VAP — Automation Artefact 4)
#
# Instructions:
#   1. Open in Google Colab (File → Upload notebook, or paste
#      this file into a new Colab notebook as a code cell)
#   2. Run the single cell — no pip installs required
#   3. Interact with the graph:
#      - Click any node  → see its details in the sidebar
#      - Drag nodes      → reposition them freely
#      - Scroll/pinch    → zoom in and out
#      - Use the toolbar → filter by Article or Relationship type
#      - Search bar      → find nodes by label or ID
# ============================================================

# ----------------------------------------------------------
# Enable the Colab custom widget manager (required for HTML
# display via IPython.display.HTML in modern Colab runtimes)
# ----------------------------------------------------------
try:
    from google.colab import output
    output.enable_custom_widget_manager()
except Exception:
    pass  # Not running in Colab — local Jupyter is fine too

from IPython.display import HTML, display

# ----------------------------------------------------------
# Graph data  (nodes + directed edges extracted from the
# original yfiles-jupyter-graphs notebook)
# ----------------------------------------------------------

NODES = [
    # --- Article 13 sub-requirements (transparency use-instructions) ---
    {"id": "ART13-1",  "label": "Identity/Contact Details for Provider",          "article": 13, "sub": True,  "url": "https://artificialintelligenceact.com/title-iii/chapter-2/article-13/", "cre": "820-877"},
    {"id": "ART13-2",  "label": "Intended Purpose",                                "article": 13, "sub": True,  "url": "https://artificialintelligenceact.com/title-iii/chapter-2/article-13/", "cre": "820-877"},
    {"id": "ART13-3",  "label": "Accuracy, Robustness & Cybersecurity Metrics",   "article": 13, "sub": True,  "url": "https://artificialintelligenceact.com/title-iii/chapter-2/article-13/", "cre": "820-877"},
    {"id": "ART13-4",  "label": "Risks of Foreseeable Misuse",                    "article": 13, "sub": True,  "url": "https://artificialintelligenceact.com/title-iii/chapter-2/article-13/", "cre": "162-655"},
    {"id": "ART13-5",  "label": "Ability to Explain AI System Output",            "article": 13, "sub": True,  "url": "https://artificialintelligenceact.com/title-iii/chapter-2/article-13/", "cre": "162-655"},
    {"id": "ART13-6",  "label": "Performance on Target Persons",                  "article": 13, "sub": True,  "url": "https://artificialintelligenceact.com/title-iii/chapter-2/article-13/", "cre": "820-877"},
    {"id": "ART13-7",  "label": "Input Data Specifications",                       "article": 13, "sub": True,  "url": "https://artificialintelligenceact.com/title-iii/chapter-2/article-13/", "cre": "820-877"},
    {"id": "ART13-8",  "label": "Info Necessary to Interpret Output",              "article": 13, "sub": True,  "url": "https://artificialintelligenceact.com/title-iii/chapter-2/article-13/", "cre": "820-877"},
    {"id": "ART13-9",  "label": "Pre-determined Conformity Changes",              "article": 13, "sub": True,  "url": "https://artificialintelligenceact.com/title-iii/chapter-2/article-13/", "cre": "820-877"},
    {"id": "ART13-10", "label": "Human Oversight Measures",                        "article": 13, "sub": True,  "url": "https://artificialintelligenceact.com/title-iii/chapter-2/article-13/", "cre": "820-877"},
    {"id": "ART13-11", "label": "Computing Resources & Maintenance",               "article": 13, "sub": True,  "url": "https://artificialintelligenceact.com/title-iii/chapter-2/article-13/", "cre": "820-877"},
    {"id": "ART13-12", "label": "Mechanisms to Collect, Store & Interpret Logs",  "article": 13, "sub": True,  "url": "https://artificialintelligenceact.com/title-iii/chapter-2/article-13/", "cre": "842-876"},

    # --- Article 16 (Provider obligations) ---
    {"id": "ART16-1",  "label": "Compliance with Ch.2 Requirements",              "article": 16, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-16/", "cre": "464-513"},
    {"id": "ART16-2",  "label": "Indicate Name, TM & Contact Address",            "article": 16, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-16/", "cre": "820-877"},
    {"id": "ART16-3",  "label": "Quality Management System",                       "article": 16, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-16/", "cre": "464-513"},
    {"id": "ART16-4",  "label": "Documentation",                                   "article": 16, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-16/", "cre": "820-877"},
    {"id": "ART16-5",  "label": "Logs",                                            "article": 16, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-16/", "cre": "842-876"},
    {"id": "ART16-6",  "label": "Conformity Assessment",                           "article": 16, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-16/", "cre": "464-513"},
    {"id": "ART16-7",  "label": "Conformity Declaration",                          "article": 16, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-16/", "cre": "464-513"},
    {"id": "ART16-8",  "label": "Affix CE Marking",                               "article": 16, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-16/", "cre": "464-513"},
    {"id": "ART16-9",  "label": "Registration",                                    "article": 16, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-16/", "cre": "510-324"},
    {"id": "ART16-10", "label": "Corrective Action / Duty of Information",        "article": 16, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-16/", "cre": "887-750"},
    {"id": "ART16-11", "label": "Regulatory Cooperation (Demonstrate Conformity)","article": 16, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-16/", "cre": "464-513"},
    {"id": "ART16-12", "label": "Accessibility Requirements",                      "article": 16, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-16/", "cre": "510-324"},

    # --- Article 17 (Quality management system contents) ---
    {"id": "ART17-1",  "label": "Strategy for Regulatory Compliance",             "article": 17, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-17/", "cre": "510-324"},
    {"id": "ART17-2",  "label": "Design, Control & Verification Measures",        "article": 17, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-17/", "cre": "326-704"},
    {"id": "ART17-3",  "label": "Development, QC & QA Measures",                  "article": 17, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-17/", "cre": "433-442"},
    {"id": "ART17-4",  "label": "Examination, Test & Validation Measures",        "article": 17, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-17/", "cre": "433-442"},
    {"id": "ART17-5",  "label": "Technical Specifications to be Applied",         "article": 17, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-17/", "cre": "787-638"},
    {"id": "ART17-6",  "label": "Data Management Systems & Procedures",           "article": 17, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-17/", "cre": "126-668"},
    {"id": "ART17-7",  "label": "Risk Management System",                          "article": 17, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-17/", "cre": "307-242"},
    {"id": "ART17-8",  "label": "Post-Market Monitoring System",                  "article": 17, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-17/", "cre": "887-750"},
    {"id": "ART17-9",  "label": "Serious Incident Reporting Procedures",          "article": 17, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-17/", "cre": "463-577"},
    {"id": "ART17-10", "label": "Communication with National Authorities",        "article": 17, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-17/", "cre": "510-324"},
    {"id": "ART17-11", "label": "Resource Management",                             "article": 17, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-17/", "cre": "563-088"},
    {"id": "ART17-12", "label": "Accountability Framework",                        "article": 17, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-17/", "cre": "427-113"},

    # --- Core Chapter 2 articles (Articles 9–15) ---
    {"id": "ART9",     "label": "Risk Management System",                          "article": 9,  "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-2/article-9/",  "cre": "307-242"},
    {"id": "ART10",    "label": "Data & Data Governance",                          "article": 10, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-2/article-10/", "cre": "287-823"},
    {"id": "ART11",    "label": "Technical Documentation",                         "article": 11, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-2/article-11/", "cre": "820-877"},
    {"id": "ART12",    "label": "Record Keeping (Logs)",                           "article": 12, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-2/article-12/", "cre": "842-876"},
    {"id": "ART13",    "label": "Transparency & Use Instructions for Deployers",   "article": 13, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-2/article-13/", "cre": "820-877"},
    {"id": "ART14",    "label": "Design/Develop to Enable Human Oversight",        "article": 14, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-2/article-14/", "cre": "433-342"},
    {"id": "ART15",    "label": "Accuracy, Robustness & Cybersecurity",            "article": 15, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-2/article-15/", "cre": "464-513"},

    # --- Article 18 (Conformity documentation) ---
    {"id": "ART18-1",  "label": "Changes Approved by Notified Bodies",            "article": 18, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-18/", "cre": "124-564"},
    {"id": "ART18-2",  "label": "Decisions/Documents by Notified Bodies",         "article": 18, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-18/", "cre": "464-513"},
    {"id": "ART18-3",  "label": "EU Declaration of Conformity",                   "article": 18, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-3/article-18/", "cre": "464-513"},

    # --- Article 51 (Registration) ---
    {"id": "ART51-1",  "label": "Register High Risk AI System Provider",           "article": 51, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-5/article-51/", "cre": "766-162"},
    {"id": "ART51-2",  "label": "Register High Risk AI System",                    "article": 51, "sub": False, "url": "https://artificialintelligenceact.com/title-iii/chapter-5/article-51/", "cre": "766-162"},
]

# Directed edges  (s = source id, t = target id, rel = relationship type)
EDGES = [
    # ART16-3 Quality Mgmt System  →  TO_INCLUDE  →  ART17-x QMS contents
    {"id":  1, "s": "ART16-3",  "t": "ART17-1",  "rel": "TO_INCLUDE"},
    {"id":  2, "s": "ART16-3",  "t": "ART17-2",  "rel": "TO_INCLUDE"},
    {"id":  3, "s": "ART16-3",  "t": "ART17-3",  "rel": "TO_INCLUDE"},
    {"id":  4, "s": "ART16-3",  "t": "ART17-4",  "rel": "TO_INCLUDE"},
    {"id":  5, "s": "ART16-3",  "t": "ART17-6",  "rel": "TO_INCLUDE"},
    {"id":  6, "s": "ART16-3",  "t": "ART17-7",  "rel": "TO_INCLUDE"},
    {"id":  7, "s": "ART16-3",  "t": "ART17-8",  "rel": "TO_INCLUDE"},
    {"id":  8, "s": "ART16-3",  "t": "ART17-9",  "rel": "TO_INCLUDE"},
    {"id":  9, "s": "ART16-3",  "t": "ART17-10", "rel": "TO_INCLUDE"},
    {"id": 10, "s": "ART16-3",  "t": "ART17-11", "rel": "TO_INCLUDE"},
    {"id": 11, "s": "ART16-3",  "t": "ART17-12", "rel": "TO_INCLUDE"},
    {"id": 12, "s": "ART16-3",  "t": "ART16-4",  "rel": "TO_INCLUDE"},
    # SEE_ALSO cross-references
    {"id": 13, "s": "ART16-10", "t": "ART17-9",  "rel": "SEE_ALSO"},
    {"id": 41, "s": "ART13-10", "t": "ART14",    "rel": "SEE_ALSO"},
    # ART16-1 Compliance  →  REQUIRES  →  Core Chapter 2 articles
    {"id": 14, "s": "ART16-1",  "t": "ART9",     "rel": "REQUIRES"},
    {"id": 15, "s": "ART16-1",  "t": "ART10",    "rel": "REQUIRES"},
    {"id": 16, "s": "ART16-1",  "t": "ART11",    "rel": "REQUIRES"},
    {"id": 17, "s": "ART16-1",  "t": "ART12",    "rel": "REQUIRES"},
    {"id": 18, "s": "ART16-1",  "t": "ART13",    "rel": "REQUIRES"},
    {"id": 19, "s": "ART16-1",  "t": "ART14",    "rel": "REQUIRES"},
    {"id": 20, "s": "ART16-1",  "t": "ART15",    "rel": "REQUIRES"},
    # RELATED_TO bidirectional semantic links
    {"id": 21, "s": "ART9",     "t": "ART17-7",  "rel": "RELATED_TO"},
    {"id": 22, "s": "ART16-4",  "t": "ART11",    "rel": "RELATED_TO"},
    {"id": 26, "s": "ART16-5",  "t": "ART12",    "rel": "RELATED_TO"},
    # ART16-4 Documentation  →  REQUIRES  →  ART18 conformity docs
    {"id": 23, "s": "ART16-4",  "t": "ART18-1",  "rel": "REQUIRES"},
    {"id": 24, "s": "ART16-4",  "t": "ART18-2",  "rel": "REQUIRES"},
    {"id": 25, "s": "ART16-4",  "t": "ART18-3",  "rel": "REQUIRES"},
    # ART16-9 Registration  →  REQUIRES  →  ART51 registrations
    {"id": 27, "s": "ART16-9",  "t": "ART51-1",  "rel": "REQUIRES"},
    {"id": 28, "s": "ART16-9",  "t": "ART51-2",  "rel": "REQUIRES"},
    # ART13 Transparency  →  REQUIRES  →  ART13 sub-requirements
    {"id": 29, "s": "ART13",    "t": "ART13-1",  "rel": "REQUIRES"},
    {"id": 30, "s": "ART13",    "t": "ART13-2",  "rel": "REQUIRES"},
    {"id": 31, "s": "ART13",    "t": "ART13-3",  "rel": "REQUIRES"},
    {"id": 32, "s": "ART13",    "t": "ART13-4",  "rel": "REQUIRES"},
    {"id": 33, "s": "ART13",    "t": "ART13-5",  "rel": "REQUIRES"},
    {"id": 34, "s": "ART13",    "t": "ART13-6",  "rel": "REQUIRES"},
    {"id": 35, "s": "ART13",    "t": "ART13-7",  "rel": "REQUIRES"},
    {"id": 36, "s": "ART13",    "t": "ART13-8",  "rel": "REQUIRES"},
    {"id": 37, "s": "ART13",    "t": "ART13-9",  "rel": "REQUIRES"},
    {"id": 38, "s": "ART13",    "t": "ART13-10", "rel": "REQUIRES"},
    {"id": 39, "s": "ART13",    "t": "ART13-11", "rel": "REQUIRES"},
    {"id": 40, "s": "ART13",    "t": "ART13-12", "rel": "REQUIRES"},
]

# ----------------------------------------------------------
# Node colour mapping  (mirrors the original notebook scheme)
# ----------------------------------------------------------
def node_color(node_id: str) -> str:
    if node_id.startswith("ART16"):   return "#e8a876"   # orange  — Art 16
    if node_id.startswith("ART17"):   return "#93e876"   # green   — Art 17
    if node_id in {"ART9","ART10","ART11","ART12","ART13","ART14","ART15"}:
                                      return "#e362d4"   # pink    — Core Ch.2
    if node_id.startswith("ART13-"): return "#ede772"   # yellow  — Art 13 subs
    if node_id.startswith("ART18"):   return "#FF6347"   # red     — Art 18
    if node_id.startswith("ART51"):   return "#76c0e8"   # blue    — Art 51
    return "#cccccc"

# Edge colour + dash pattern
EDGE_COLOR = {
    "REQUIRES":   "#4a9eff",
    "RELATED_TO": "#22c97a",
    "TO_INCLUDE": "#f5a623",
    "SEE_ALSO":   "#9c9a92",
}
EDGE_DASH = {
    "REQUIRES":   "none",
    "RELATED_TO": "none",
    "TO_INCLUDE": "none",
    "SEE_ALSO":   "5,4",
}

# ----------------------------------------------------------
# Build the self-contained HTML  (D3 v7 force-directed graph
# with interactive sidebar, article/relation filter pills,
# live search, zoom/pan, and drag-to-reposition nodes)
# ----------------------------------------------------------

import json

nodes_json = json.dumps(
    [{"id": n["id"], "label": n["label"], "article": n["article"],
      "sub": n.get("sub", False), "url": n["url"], "cre": n["cre"],
      "color": node_color(n["id"])}
     for n in NODES]
)

edges_json = json.dumps(
    [{"id": e["id"], "source": e["s"], "target": e["t"],
      "rel": e["rel"],
      "color": EDGE_COLOR[e["rel"]],
      "dash":  EDGE_DASH[e["rel"]]}
     for e in EDGES]
)

articles_json  = json.dumps(sorted({n["article"] for n in NODES}))
relations_json = json.dumps(sorted({e["rel"]     for e in EDGES}))

# Pill background colours keyed by article number
art_pill_bg = {
    9: "#e362d4", 10: "#e362d4", 11: "#e362d4",
    12: "#e362d4", 13: "#e362d4", 14: "#e362d4", 15: "#e362d4",
    16: "#e8a876", 17: "#93e876", 18: "#FF6347", 51: "#76c0e8",
}
art_pill_json = json.dumps({str(k): v for k, v in art_pill_bg.items()})

rel_pill_json = json.dumps(EDGE_COLOR)

HTML_TEMPLATE = """
<!DOCTYPE html>
<html lang="en">
<head>
<meta charset="UTF-8">
<meta name="viewport" content="width=device-width, initial-scale=1.0">
<title>EU AI Act Requirements Graph</title>
<style>
  *, *::before, *::after {{ box-sizing: border-box; margin: 0; padding: 0; }}
  body {{ font-family: system-ui, -apple-system, sans-serif; background: #fff; color: #111; }}
  #wrap {{ display: flex; flex-direction: column; width: 100%; max-width: 1100px; margin: 0 auto; border: 1px solid #e0e0e0; border-radius: 10px; overflow: hidden; }}
  h1 {{ font-size: 15px; font-weight: 600; padding: 10px 14px 0; color: #111; }}
  p.sub {{ font-size: 11px; color: #666; padding: 2px 14px 8px; }}
  #toolbar {{ display: flex; flex-wrap: wrap; gap: 8px; padding: 8px 12px; border-bottom: 1px solid #e8e8e8; align-items: center; background: #f7f7f5; }}
  #search {{ flex: 1; min-width: 130px; max-width: 240px; height: 30px; font-size: 13px; padding: 0 10px; border: 1px solid #d0d0d0; border-radius: 6px; outline: none; }}
  #search:focus {{ border-color: #888; }}
  .ctl-label {{ font-size: 11px; color: #666; white-space: nowrap; }}
  .pill-group {{ display: flex; gap: 4px; flex-wrap: wrap; }}
  .pill {{ font-size: 11px; padding: 3px 9px; border-radius: 99px; border: 1px solid transparent; cursor: pointer; user-select: none; transition: opacity .15s; }}
  .pill.off {{ opacity: .3; }}
  #canvas-wrap {{ position: relative; overflow: hidden; }}
  svg#g {{ display: block; width: 100%; cursor: grab; }}
  svg#g:active {{ cursor: grabbing; }}
  .link {{ stroke-opacity: .6; transition: stroke-opacity .2s; }}
  .node circle {{ cursor: pointer; transition: r .15s; }}
  .node circle:hover {{ opacity: .85; }}
  .node text {{ pointer-events: none; font-family: system-ui, sans-serif; dominant-baseline: central; }}
  #side {{ position: absolute; top: 0; right: 0; width: 240px; height: 100%; background: #f9f9f7; border-left: 1px solid #e0e0e0; padding: 14px; overflow-y: auto; transition: transform .25s; display: flex; flex-direction: column; gap: 8px; }}
  #side.hidden {{ transform: translateX(100%); }}
  #side h3 {{ font-size: 13px; font-weight: 600; color: #111; line-height: 1.4; }}
  .meta {{ font-size: 11px; color: #555; line-height: 1.7; }}
  .art-badge {{ display: inline-block; font-size: 11px; font-weight: 600; padding: 2px 8px; border-radius: 99px; margin-bottom: 4px; }}
  .side-link {{ font-size: 11px; color: #2563eb; text-decoration: none; }}
  .side-link:hover {{ text-decoration: underline; }}
  .neighbors {{ font-size: 11px; color: #555; border-top: 1px solid #e8e8e8; padding-top: 8px; }}
  .neighbors b {{ color: #111; }}
  .nb-item {{ cursor: pointer; padding: 2px 0; }}
  .nb-item:hover {{ text-decoration: underline; }}
  #close-btn {{ cursor: pointer; font-size: 11px; color: #888; align-self: flex-end; margin-bottom: -4px; }}
  #close-btn:hover {{ color: #111; }}
  #legend {{ display: flex; flex-wrap: wrap; gap: 6px 14px; padding: 8px 12px; border-top: 1px solid #e8e8e8; background: #f7f7f5; }}
  .leg-item {{ display: flex; align-items: center; gap: 5px; font-size: 11px; color: #555; }}
  .leg-dot {{ width: 10px; height: 10px; border-radius: 50%; flex-shrink: 0; }}
  .leg-line {{ width: 18px; height: 2px; flex-shrink: 0; border-radius: 1px; }}
  #stats {{ font-size: 11px; color: #999; padding: 4px 12px; border-top: 1px solid #e8e8e8; background: #f7f7f5; }}
  #reset-btn {{ font-size: 11px; padding: 3px 9px; border-radius: 6px; border: 1px solid #d0d0d0; background: transparent; cursor: pointer; color: #555; }}
  #reset-btn:hover {{ background: #eee; }}
</style>
</head>
<body>
<div id="wrap">
  <h1>EU AI Act — Requirements for Providers of High Risk AI Systems</h1>
  <p class="sub">Automation Artefact · AI-Based Phishing &amp; Email Security Systems (Proofpoint TAP/VAP)</p>
  <div id="toolbar">
    <input type="text" id="search" placeholder="Search nodes…" oninput="doSearch(this.value)">
    <span class="ctl-label">Article:</span>
    <div class="pill-group" id="art-pills"></div>
    <span class="ctl-label">Relation:</span>
    <div class="pill-group" id="rel-pills"></div>
    <button id="reset-btn" onclick="resetView()">Reset</button>
  </div>
  <div id="canvas-wrap">
    <svg id="g"></svg>
    <div id="side" class="hidden">
      <span id="close-btn" onclick="closeSide()">✕ close</span>
      <h3 id="s-title"></h3>
      <div class="meta" id="s-meta"></div>
      <a id="s-link" href="#" target="_blank" class="side-link" style="display:none">Open EU AI Act text ↗</a>
      <div class="neighbors" id="s-neighbors"></div>
    </div>
  </div>
  <div id="legend">
    <span class="ctl-label">Node:</span>
    <div class="leg-item"><div class="leg-dot" style="background:#e8a876"></div>Art 16 (Provider obligations)</div>
    <div class="leg-item"><div class="leg-dot" style="background:#93e876"></div>Art 17 (Quality mgmt)</div>
    <div class="leg-item"><div class="leg-dot" style="background:#e362d4"></div>Arts 9–15 (Core HR requirements)</div>
    <div class="leg-item"><div class="leg-dot" style="background:#ede772"></div>Art 13 sub-requirements</div>
    <div class="leg-item"><div class="leg-dot" style="background:#FF6347"></div>Art 18 (Conformity docs)</div>
    <div class="leg-item"><div class="leg-dot" style="background:#76c0e8"></div>Art 51 (Registration)</div>
    <span class="ctl-label" style="margin-left:6px">Edge:</span>
    <div class="leg-item"><div class="leg-line" style="background:#4a9eff"></div>REQUIRES</div>
    <div class="leg-item"><div class="leg-line" style="background:#22c97a"></div>RELATED_TO</div>
    <div class="leg-item"><div class="leg-line" style="background:#f5a623"></div>TO_INCLUDE</div>
    <div class="leg-item"><div class="leg-line" style="background:#9c9a92;background:repeating-linear-gradient(90deg,#9c9a92 0,#9c9a92 4px,transparent 4px,transparent 7px)"></div>SEE_ALSO</div>
  </div>
  <div id="stats">Loading…</div>
</div>

<script src="https://cdnjs.cloudflare.com/ajax/libs/d3/7.8.5/d3.min.js"></script>
<script>
const NODES_DATA   = {nodes_json};
const EDGES_DATA   = {edges_json};
const ARTICLES     = {articles_json};
const RELATIONS    = {relations_json};
const ART_PILL_BG  = {art_pill_json};
const REL_EDGE_COL = {rel_pill_json};

const EDGE_DASH = {{ REQUIRES:"none", RELATED_TO:"none", TO_INCLUDE:"none", SEE_ALSO:"5,4" }};

let activeArts = new Set(ARTICLES);
let activeRels = new Set(RELATIONS);
let searchTerm = "";
let selectedId = null;
let sim, linkSel, nodeSel;

// ---- Build filter pills ----
const artPillsEl = document.getElementById("art-pills");
ARTICLES.forEach(a => {{
  const bg = ART_PILL_BG[String(a)] || "#ccc";
  const el = document.createElement("span");
  el.className = "pill";
  el.textContent = "Art " + a;
  el.style.background = bg + "33";
  el.style.borderColor = bg + "88";
  el.onclick = () => {{
    activeArts.has(a) ? activeArts.delete(a) : activeArts.add(a);
    el.classList.toggle("off", !activeArts.has(a));
    update();
  }};
  artPillsEl.appendChild(el);
}});

const relPillsEl = document.getElementById("rel-pills");
RELATIONS.forEach(r => {{
  const col = REL_EDGE_COL[r] || "#ccc";
  const el = document.createElement("span");
  el.className = "pill";
  el.textContent = r.replace(/_/g," ").toLowerCase();
  el.style.background = col + "22";
  el.style.borderColor = col + "66";
  el.onclick = () => {{
    activeRels.has(r) ? activeRels.delete(r) : activeRels.add(r);
    el.classList.toggle("off", !activeRels.has(r));
    update();
  }};
  relPillsEl.appendChild(el);
}});

// ---- SVG + zoom ----
const W = 900, H = 560;
const svg = d3.select("#g").attr("viewBox", `0 0 ${{W}} ${{H}}`).attr("preserveAspectRatio","xMidYMid meet");
svg.node().style.height = H + "px";

const defs = svg.append("defs");
RELATIONS.forEach(rel => {{
  const col = REL_EDGE_COL[rel];
  const m = defs.append("marker")
    .attr("id","arr-"+rel)
    .attr("viewBox","0 0 10 10").attr("refX",9).attr("refY",5)
    .attr("markerWidth",5).attr("markerHeight",5).attr("orient","auto");
  m.append("path").attr("d","M2 1L8 5L2 9").attr("fill","none")
    .attr("stroke",col).attr("stroke-width",1.6)
    .attr("stroke-linecap","round").attr("stroke-linejoin","round");
}});

const g = svg.append("g");
const zoom = d3.zoom().scaleExtent([0.2, 3]).on("zoom", e => g.attr("transform", e.transform));
svg.call(zoom);

// ---- Simulation ----
const nodesData = NODES_DATA.map(n => ({{...n}}));
const edgesData = EDGES_DATA.map(e => ({{...e}}));

function nodeRadius(n) {{
  const deg = edgesData.filter(e => e.source?.id === n.id || e.target?.id === n.id ||
                                     e.source === n.id     || e.target === n.id).length;
  return Math.max(9, Math.min(16, 7 + deg));
}}

sim = d3.forceSimulation(nodesData)
  .force("link", d3.forceLink(edgesData).id(d => d.id)
    .distance(d => d.rel==="TO_INCLUDE" ? 70 : d.rel==="REQUIRES" ? 85 : 100)
    .strength(0.5))
  .force("charge", d3.forceManyBody().strength(-230))
  .force("center",  d3.forceCenter(W/2, H/2))
  .force("collision", d3.forceCollide(22))
  .on("tick", ticked);

linkSel = g.append("g").selectAll("line");
nodeSel = g.append("g").selectAll("g.node");

// ---- Update (called on filter/search change) ----
function visibleNodes() {{
  return nodesData.filter(n =>
    activeArts.has(n.article) &&
    (!searchTerm || n.label.toLowerCase().includes(searchTerm) || n.id.toLowerCase().includes(searchTerm))
  );
}}

function visibleEdges(vn) {{
  const ids = new Set(vn.map(n => n.id));
  return edgesData.filter(e => {{
    const sid = e.source?.id ?? e.source;
    const tid = e.target?.id ?? e.target;
    return ids.has(sid) && ids.has(tid) && activeRels.has(e.rel);
  }});
}}

function update() {{
  const vn = visibleNodes();
  const ve = visibleEdges(vn);

  // Links
  linkSel = g.select("g:first-child").selectAll("line").data(ve, d => d.id);
  linkSel.exit().remove();
  const lEnter = linkSel.enter().append("line").attr("class","link")
    .attr("stroke", d => d.color)
    .attr("stroke-width", d => d.rel==="REQUIRES" ? 1.8 : 1.2)
    .attr("stroke-dasharray", d => EDGE_DASH[d.rel] || "none")
    .attr("marker-end", d => "url(#arr-"+d.rel+")");
  linkSel = lEnter.merge(linkSel);

  // Nodes
  nodeSel = g.select("g:last-child").selectAll("g.node").data(vn, d => d.id);
  nodeSel.exit().remove();
  const nEnter = nodeSel.enter().append("g").attr("class","node")
    .call(d3.drag().on("start", dragStart).on("drag", dragging).on("end", dragEnd))
    .on("click", (e, d) => {{ e.stopPropagation(); selectNode(d.id); }});
  nEnter.append("circle")
    .attr("r", d => nodeRadius(d))
    .attr("fill", d => d.color)
    .attr("stroke","#fff").attr("stroke-width",1.5);
  nEnter.append("text")
    .attr("dy","2.2em").attr("text-anchor","middle")
    .style("font-size","9px").style("fill","#666")
    .text(d => d.id.replace("ART",""));
  nodeSel = nEnter.merge(nodeSel);
  nodeSel.select("circle").attr("fill", d => d.color).attr("r", d => nodeRadius(d));

  sim.nodes(vn);
  sim.force("link").links(ve);
  sim.alpha(0.4).restart();

  document.getElementById("stats").textContent =
    `Showing ${{vn.length}} of ${{NODES_DATA.length}} nodes · ${{ve.length}} of ${{EDGES_DATA.length}} edges`;
}}

function ticked() {{
  if (linkSel) linkSel
    .attr("x1", d => d.source?.x ?? 0).attr("y1", d => d.source?.y ?? 0)
    .attr("x2", d => d.target?.x ?? 0).attr("y2", d => d.target?.y ?? 0);
  if (nodeSel) nodeSel.attr("transform", d => `translate(${{d.x??0}},${{d.y??0}})`);
}}

svg.on("click", () => {{ closeSide(); selectedId = null; }});

// ---- Node detail sidebar ----
function selectNode(id) {{
  selectedId = id;
  const n = NODES_DATA.find(x => x.id === id);
  if (!n) return;
  const artNames = {{9:"Risk Mgmt",10:"Data Governance",11:"Tech Documentation",12:"Record Keeping",
    13:"Transparency",14:"Human Oversight",15:"Accuracy & Cybersec",16:"Provider Obligations",
    17:"Quality Mgmt System",18:"Conformity Docs",51:"Registration"}};
  document.getElementById("s-title").textContent = n.label;
  document.getElementById("s-meta").innerHTML =
    `<span class="art-badge" style="background:${{n.color}}33;border:1px solid ${{n.color}}88">` +
    `Art. ${{n.article}}${{n.sub?" (sub-req)":""}}</span><br>` +
    `${{artNames[n.article]||""}}<br>CRE: ${{n.cre||"—"}}`;
  const linkEl = document.getElementById("s-link");
  if (n.url) {{ linkEl.href = n.url; linkEl.style.display = "block"; }}
  else        {{ linkEl.style.display = "none"; }}

  const outgoing = EDGES_DATA.filter(e => (e.source?.id ?? e.source) === id);
  const incoming = EDGES_DATA.filter(e => (e.target?.id ?? e.target) === id);
  let html = "";
  if (outgoing.length) {{
    html += `<b>Links to (${{outgoing.length}}):</b><br>`;
    outgoing.forEach(e => {{
      const tid = e.target?.id ?? e.target;
      const tn  = NODES_DATA.find(x => x.id === tid);
      html += `<div class="nb-item" onclick="selectNode('${{tid}}')" style="color:${{e.color}}">` +
              `<span style="opacity:.7">${{e.rel.replace(/_/g," ").toLowerCase()}} → </span>` +
              `${{tn ? tn.label : tid}}</div>`;
    }});
  }}
  if (incoming.length) {{
    html += `<br><b>Linked from (${{incoming.length}}):</b><br>`;
    incoming.forEach(e => {{
      const sid = e.source?.id ?? e.source;
      const sn  = NODES_DATA.find(x => x.id === sid);
      html += `<div class="nb-item" onclick="selectNode('${{sid}}')" style="color:${{e.color}}">` +
              `<span style="opacity:.7">← ${{e.rel.replace(/_/g," ").toLowerCase()}} </span>` +
              `${{sn ? sn.label : sid}}</div>`;
    }});
  }}
  document.getElementById("s-neighbors").innerHTML = html || "<span>No visible connections</span>";
  document.getElementById("side").classList.remove("hidden");

  nodeSel.select("circle")
    .attr("stroke-width", d => d.id === id ? 3 : 1.5)
    .attr("stroke",       d => d.id === id ? "#111" : "#fff");
  linkSel.attr("stroke-opacity", e => {{
    const sid = e.source?.id ?? e.source;
    const tid = e.target?.id ?? e.target;
    return (sid === id || tid === id) ? 1 : 0.1;
  }});
}}

function closeSide() {{
  document.getElementById("side").classList.add("hidden");
  if (nodeSel) nodeSel.select("circle").attr("stroke-width",1.5).attr("stroke","#fff");
  if (linkSel) linkSel.attr("stroke-opacity", 0.6);
}}

function doSearch(v) {{ searchTerm = v.trim().toLowerCase(); update(); }}

function resetView() {{
  svg.transition().duration(400).call(zoom.transform, d3.zoomIdentity);
  closeSide();
  activeArts = new Set(ARTICLES);
  activeRels = new Set(RELATIONS);
  searchTerm = "";
  document.getElementById("search").value = "";
  document.querySelectorAll(".pill").forEach(p => p.classList.remove("off"));
  update();
}}

function dragStart(e, d) {{ if (!e.active) sim.alphaTarget(0.3).restart(); d.fx = d.x; d.fy = d.y; }}
function dragging(e, d)   {{ d.fx = e.x; d.fy = e.y; }}
function dragEnd(e, d)    {{ if (!e.active) sim.alphaTarget(0); d.fx = null; d.fy = null; }}

// Initial render
update();
</script>
</body>
</html>
""".format(
    nodes_json=nodes_json,
    edges_json=edges_json,
    articles_json=articles_json,
    relations_json=relations_json,
    art_pill_json=art_pill_json,
    rel_pill_json=rel_pill_json,
)

# ----------------------------------------------------------
# Render in the Colab / Jupyter output cell
# ----------------------------------------------------------
display(HTML(HTML_TEMPLATE))
```
